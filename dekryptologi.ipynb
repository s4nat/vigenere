```python
coincidences = []

for i in range(1, 100):
    coincidence_count = 0
    for j in range(len(ciphertext) - i):
        if ciphertext[j] == ciphertext[j + i]:
            coincidence_count += 1
    coincidences.append(coincidence_count)


print(coincidences)
```

```python
plt.plot(coincidences)
```

```python
sum = 0
for coincidence in coincidences:
    sum += coincidence

mean = sum / len(coincidences)
mean
```
